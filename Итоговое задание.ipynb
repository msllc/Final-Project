# Динамика по зарплатам с учетом инфляции 

```python
inflation = pd.read_csv(r'C:\Users\Nastya\Downloads\inflation_.csv')
```

Оставим только данные с 2000 по 2023 года (в соответствии с имеющимися данными по зарплатам)

```python
inflation = inflation.loc[1:24]
inflation = inflation.loc[::-1]
inflation.index = inflation['Год']
inflation.drop(labels = 'Год', axis = 1, inplace = True)
```

```python
for i in ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек', 'Всего']:
    inflation[i] = inflation[i].apply(lambda x: x.replace(',', '.')).astype('float')
```

```python
data = data.T
```

Посчитаем показатели с учетом инфляции

```python
data_ = data.copy()
```

```python
for j in range(2000, 2023):
    for i in range(j, 2023):
        data_.loc[j] = data_.loc[j]*((inflation['Всего'].loc[i]/(100)) + 1)
```

```python
plt.figure(figsize=(18,6))
for i in data_.columns:
    plt.plot(data_[i], label=i)
plt.title('Изменение зарплаты по годам по видам экономической деятельности учетом инфляции, руб.')
plt.legend()
plt.show()
```

Приведя значения ЗП к более акутальным значениям, мы замечаем увеличение флуктуации на графиках. В отрасли государственного управления пик ЗП приходится на 2013 год. 

# Сравнение номинальных и реальных ЗП по отраслям

```python
plt.figure(figsize=(18,6))
plt.plot(data_['Добыча полезных ископаемых'], label='Добыча полезных икопаемых с учетом инфляции')
plt.plot(data['Добыча полезных ископаемых'], label='Добыча полезных икопаемых')
plt.title('Изменение ЗП в сфере добычи полезных ископаемых')
plt.legend()
plt.show()
```

ЗП в сфере добычи полезных ископаемых в целом показывает стабильный рост, за исключением 2009, 2015, 2016 гг. в реальном выражении.

```python
plt.figure(figsize=(18,6))
plt.plot(data_['Гос. управление, обеспечение ВБ, социальное обеспечение'], label='С учетом инфляции')
plt.plot(data['Гос. управление, обеспечение ВБ, социальное обеспечение'], label='Без учета инфляции')
plt.title('Изменение ЗП в сфере гос.управления, обеспечения ВБ и соц. обеспечения')
plt.legend()
plt.show()
```

ЗП в сфере гос. управления, обеспечения ВБ и соц. обеспечения, по наиболее свежим данным, не превышают пиковых значений 2012-2014 гг., но при этом присутствует тенденция к росту. 

```python
plt.figure(figsize=(18,6))
plt.plot(data_['Здравоохранение и соц. услуги'], label='С учетом инфляции')
plt.plot(data['Здравоохранение и соц. услуги'], label='Без учета инфляции')
plt.title('Изменение ЗП в сфере здравоохранения и соц. услуг')
plt.legend()
plt.show()
```

Пик зарплат в сфере здравоохранения и соц. услуг в реальных значениях пришелся на 2020 год, в последнее время значения ЗП несколько снизились
